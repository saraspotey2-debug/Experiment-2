```python
#Creating a list and printing it (where elements are duplicated).
listfruits=["apple","orange","apple","cherry"]
print(listfruits)
```

```python
# Creating a list and printing the list and lenght of the list.
list1=[2,2.5,True,"sara"]
print(list1)
print(len(list1))
```

```python
# Printing the elements by indexing and appending the element at back in the list.
listfruits=["apple","orange","cherry"]
print(listfruits[0])
print(listfruits[-2])
listfruits.append("banana")
print(listfruits[0])
print(listfruits[-2])
print(listfruits)
```

```python
# Inserting element by indexing at a position
listfruits=["apple","orange","cherry"]
listfruits.insert(1,"banana")
print(listfruits)
```

```python
#Interchanging the element
listfruits=["apple","orange","cherry"]
listfruits[1]="banana"
print(listfruits)
```

```python
#Combining two lists
list1=["apple","orange","cherry"]
list2=["mango","pineapple","banana"]
list1.extend(list2)
print(list1)
```

```python
# Printing specific elements by indexing
list1=["apple","banana","cherry","kiwi","mango","pineapple"]
print(list1[2:5])
print(list1[2:])
print(list1[:4])
```
